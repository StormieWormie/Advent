```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
grid = np.full(shape = (1000,1000),fill_value=False)

file = open('input.txt')
for line in file.readlines():
    info = line.split()
    if info[0] == 'turn':
        command = info[1]
        info = info[2::]
    else:
        command = info[0]
        info = info[1::]
    
    start = np.array(info[0].split(','),dtype=int) 
    end = np.array(info[2].split(','),dtype=int) 
    
    match command:
        case 'on':
            grid[start[0]:end[0]+1,start[1]:end[1]+1] = True
        case 'off':
            grid[start[0]:end[0]+1,start[1]:end[1]+1] = False
        case 'toggle':
            grid[start[0]:end[0]+1,start[1]:end[1]+1] = ~grid[start[0]:end[0]+1,start[1]:end[1]+1]
np.sum(grid)
```

```python
grid = np.full(shape = (1000,1000),fill_value=0,dtype=int)

file = open('input.txt')
for line in file.readlines():
    info = line.split()
    if info[0] == 'turn':
        command = info[1]
        info = info[2::]
    else:
        command = info[0]
        info = info[1::]
    
    start = np.array(info[0].split(','),dtype=int) 
    end = np.array(info[2].split(','),dtype=int) 
    
    match command:
        case 'on':
            grid[start[0]:end[0]+1,start[1]:end[1]+1] += 1
        case 'off':
            grid[start[0]:end[0]+1,start[1]:end[1]+1] -= 1
            grid[grid < 0] = 0
        case 'toggle':
            grid[start[0]:end[0]+1,start[1]:end[1]+1] += 2
np.sum(grid)
```
