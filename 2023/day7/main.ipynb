```python
filename = "input"

file = open(filename+'.txt')

def find_label(string):
    record = {}
    for char in string:
        if char not in record:
            record[char] = 1
        else:
            record[char] +=1
    freq = [record[key] for key in record.keys()]
    match max(freq):
        case 5:
            return "Five of a kind"
        case 4:
            return "Four of a kind"
        case 3:
            match min(freq):
                case 2:
                    return "Full house"
                case 1:
                    return "Three of a kind"
        case 2:
            match len(freq):
                case 3:
                    return "Two pair"
                case 4:
                    return "One pair"
        case 1:
            return "High card"

order = ["Five of a kind",
         "Four of a kind",
         "Full house",
         "Three of a kind",
         "Two pair",
         "One pair",
         "High card"]

types = ["A","K","Q","J","T"] + [str(i) for i in range(9,1,-1)]
value = {}
val = 0
for t in types:
    value[t] = val
    val += 1

data = {}   

for line in file.readlines():
    info = line.split()
    hand = info[0]
    bid = int(info[1])
    label = find_label(hand)

    if label not in data:
        data[label] = []

    data[label].append((hand,bid))

result = 0

def compare(hand1,hand2):
    if hand1 == hand2:
        print("oops, double")
        return False
    for char1,char2 in zip(hand1,hand2):
        if char1 == char2:
            pass
        else:
            return True if value[char1] < value[char2] else False

ordered_list = []


for label in order:
    if label in data:
        new_list = []
        for handbid in data[label]:
            N = len(new_list)
            I = None
            for i in range(N):
                if compare(handbid[0],new_list[i][0]):
                    I = i
                    break
            if I == None:
                new_list.append(handbid)
            else:
                new_list.insert(I,handbid)
        data[label] = new_list

rank = 1
result = 0
for label in reversed(order):
    if label in data:
        for handbid in reversed(data[label]):
            result += rank * handbid[1]
            rank += 1
print(result)
            
```
