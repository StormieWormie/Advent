```python
filename = "input"
file = open(filename+".txt")

result = 0

def grow(X):
    length = len(X)
    for i in range(length):
        if X[i]-1 not in X:
            X.append(X[i]-1)
        if X[i]+1 not in X:
            X.append(X[i]+1)
    X = np.sort(X)
    return list(X)

def shrink(X,bool_list):
    Y = []
    for i in X:
        if i < len(bool_list):
            if bool_list[i]:
                Y.append(i)
    return Y

def splitter(X):
    if len(X) == 0:
        return []
    Y = [[X[0]]]
    if len(X) == 1:
        return Y
    I = 0
    for i in X[1::]:
        if i-1 in Y[I]:
            Y[I].append(i)
        else:
            Y.append([i])
            I+=1
    return Y

line1 = file.readline().removesuffix('\n')
line2 = file.readline().removesuffix('\n')
state1 = np.where([i not in ignore for i in line1])[0]
state2 = np.where([i not in ignore for i in line2])[0]


numbered = [i in numbers for i in line1]
active = list(np.where([numbered[i] and
                        (((i-1 in state1 or i+1 in state1) if len(state1)>0 else False) or
                         ((i-1 in state2 or i in state2 or i+1 in state2) if len(state2)>0 else False)) for i in range(len(line1))])[0])

expanding = True
while expanding:
    start_len = len(active)
    active = grow(active)
    active = shrink(active, numbered)
    end_len = len(active)
    expanding = start_len is not end_len

if len(active)>0:
    sequences = splitter(active)
    for seq in sequences:
        word = ""
        for i in seq:
            word += line1[i]
        result += int(word)

# print(line1)
# print(active)

for line in file.readlines():
    line3 = line.removesuffix('\n')
    state3 = np.where([i not in ignore for i in line3])[0]
    active3 = [i in numbers for i in line3]

    numbered = [i in numbers for i in line2]
    active = list(np.where([numbered[i] and 
                            (((i-1 in state1 or i in state1 or i+1 in state1) if len(state1)>0 else False) or 
                             ((i-1 in state2 or i+1 in state2) if len(state2)>0 else False) or
                             ((i-1 in state3 or i in state3 or i+1 in state3) if len(state3)>0 else False)) for i in range(len(line2))])[0])
    expanding = True
    while expanding:
        start_len = len(active)
        active = grow(active)
        active = shrink(active, numbered)
        end_len = len(active)
        expanding = start_len is not end_len
    
    if len(active)>0:
        sequences = splitter(active)
        for seq in sequences:
            word = ""
            for i in seq:
                word += line2[i]
            result += int(word)
    # print(line1,line2,line3)
    # print(state1,state2,state3)
    # print("active:", active)
    # print()
    line1, state1 = line2, state2
    line2, state2 = line3, state3

numbered = [i in numbers for i in line2]
active = list(np.where([numbered[i] and
                        (((i-1 in state1 or i+1 in state2) if len(state2)>0 else False) or
                         ((i-1 in state1 or i in state1 or i+1 in state1) if len(state1)>0 else False)) for i in range(len(line1))])[0])
expanding = True
while expanding:
    start_len = len(active)
    active = grow(active)
    active = shrink(active, numbered)
    end_len = len(active)
    expanding = start_len is not end_len

if len(active)>0:
    sequences = splitter(active)
    for seq in sequences:
        word = ""
        for i in seq:
            word += line2[i]
        result += int(word)
# print(line1,line2)
# print(state1,state2)
# print("active:", active)

print(result)
```
